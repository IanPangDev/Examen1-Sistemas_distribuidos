```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
df = pd.read_csv('salidas.csv', header=None)
df.head()
```

```python
color_map = {'S': 'lightgreen', 'T': 'lightblue', 'P': 'lightcoral'}

fig, ax = plt.subplots()
table = ax.table(cellText=df.values, rowLabels=[f'Ejecución {i+1}' for i in range(df.shape[0])], loc='center')

for (i, j), cell in table.get_celld().items():
    if j >= 0:  # Excluimos las cabeceras de fila y columna
        value = df.iloc[i, j]  # Los índices en la tabla comienzan en 1
        cell.set_facecolor(color_map.get(value, 'white'))  # Asignar color basado en el valor

# Ajustar propiedades de la tabla
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 2) 

legend_labels = [Patch(color=color_map['S'], label='Segundo hilo'),
                Patch(color=color_map['T'], label='Tercer hilo'),
                Patch(color=color_map['P'], label='Primer hilo')]
plt.legend(handles=legend_labels, loc='upper center')

# Eliminar los ejes
plt.axis('off')
plt.show()
```
